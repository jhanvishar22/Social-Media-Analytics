```python
reviewan = []
for i in range(0,len(data)):
    review = re.sub('[^a-zA-Z]',' ',data['Review'][i])
    review = review.lower()
    review = review.split()
    leme = WordNetLemmatizer()
    review = [leme.lemmatize(word) for word in review]
    review = [word for word in review if not word in set(stop_words)]
    review = ' '.join(review)
    reviewan.append(review)
reviewan
```

```python
import gensim

from gensim.utils import simple_preprocess
```

```python
def sent_to_words(reviewan):
    for sentence in reviewan:
        yield(gensim.utils.simple_preprocess(str(reviewan), deacc=True))




def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc))

          if word not in stop_words] for doc in texts]



data = data.Review.values.tolist()

print(data)

data_words = list(sent_to_words(data))

# remove stop words

data_words = remove_stopwords(data_words)

print(data_words[:1][0][:30])
```

```python
import gensim.corpora as corpora

# Create Dictionary

id2word = corpora.Dictionary(data_words)

# Create Corpus

texts = data_words

# Term Document Frequency

corpusan = [id2word.doc2bow(text) for text in texts]

# View

print(reviewan[:1][0][:30])
```

```python
from pprint import pprint

# number of topics
num_topics = 10

# Build LDA model
lda_model = gensim.models.LdaMulticore(reviewan=reviewan,

                                       id2word=id2word,

                                       num_topics=num_topics)

# Printing the Keyword in the 10 topics
pprint(lda_model.print_topics())

doc_lda = lda_model[reviewan]
```
